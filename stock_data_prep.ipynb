```python
import os
import re
import pandas as pd
import numpy as np
```

```python
# initializing file path
path = './stocknet-dataset/price/raw'
```

```python
# declaring an empty dictionary
dct = {
    'Company': [],
    'Date': [],
    'Open': [],
    'High': [],
    'Low':[],
    'Close':[],
    'Adj Close':[],
    'Volume':[]
}

# helper function to add value to a column in dictionary
def addEntry(d, k, v):
    tmp = d.get(k)
    tmp.append(v)
    d[k] = tmp
```

```python
companies = os.listdir(path+'/')
for company in companies:
    data = open(path+'/'+company, 'r')
    text = data.read().split('\n')[:-1]
    text.pop(0)
    for line in text:
        addEntry(dct, 'Company', company.removesuffix('.csv'))
        sline = line.split(',')
        addEntry(dct, 'Date', sline[0])
        addEntry(dct, 'Open', sline[1])
        addEntry(dct, 'High', sline[2])
        addEntry(dct, 'Low', sline[3])
        addEntry(dct, 'Close', sline[4])
        addEntry(dct, 'Adj Close', sline[5])
        addEntry(dct, 'Volume', sline[6])
```
